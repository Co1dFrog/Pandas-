## 1.1 Pandas简介

Pandas 代表“Python Data Analysis Library”。根据 Pandas 的维基百科页面，该名称源自术语“Panel Data”，这是多维结构化数据集的计量经济学术语。但我认为Pandas只是一个超级有用的 Python 库的“Cute Name”

在使用 Python 分析数据时，Pandas 是一个相当强劲工具，即使不是最常用的工具，它也是数据处理/争论中最受欢迎和广泛使用的工具之一。 Pandas 是开源的，免费使用（在 BSD 许可下），最初由 Wes McKinney 编写。

Pandas 最酷的地方在于它获取数据（如 CSV 、HTML 文件，或 SQL 数据库等）并创建一个 Python 对象，其中包含称为数据框的行和列，看起来非常类似于统计软件中的表格（想想 Excel 或 SPSS例如。熟悉 R 的人也会看到与 R 的相似之处）。与通过 for 循环或列表理解处理列表和/或字典相比，Pandas更容易使用。

## 1.3 对Pandas DataFrame的CRUD操作

CRUD，全称增加（Create，意为“创建”）、删除（Delete）、查询（Read，意为“读取”）、改正（Update，意为“更新”），在计算机程序语言中是一连串常见的动作行为，而其行为通常是为了针对某个特定资源所作出的举动（例如：创建资料、读取资料等）。这四个行为最常见的用途是在使用SQL数据库与网站的API端口口的时候。这个词语（CRUD）最早被记载于James Martin所撰写的Managing the Data-base Environment书中。

在不同的语境下，CRUD略有差异，话虽如此，虽然有着不同名称，但底层的概念都基本相同。
